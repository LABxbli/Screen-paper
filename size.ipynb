```python
%%time
############Crop the 384-array image into individual color blocks for each mutant.
#The cutting step is not mandatory, but it facilitates the subsequent review of the actual images corresponding to the identified mutants. Therefore, we perform this cutting step.
import os 
import cv2
import numpy as np
import pylab
import matplotlib.pyplot as plt

def file_name2(file_dir): 
    L=[] 
    for root, dirs, files in os.walk(file_dir):
        for file in files:
            if os.path.splitext(file)[1] == '.png':
                L.append(os.path.join(root, file))
    return L

def cv_imread(filePath):#
    cv_img=cv2.imdecode(np.fromfile(filePath,dtype=np.uint8),-1)
    return cv_img 

def divide_spot(file_name):
    img = cv_imread(file_name)
    file_name=file_name.split(".png")[0].replace("âˆ’","-")
    os.mkdir(file_name) 
    #print(":",img.shape)
    w=img.shape[1]
    h=img.shape[0]
    h0=h/16
    w0=w/24   
    j=0
    while j<16:
        i=0
        while i<24:
            img_names=file_name+"/"+str(24*j+i+1)+".png"
            a=round(j*h0)##Rounding may sometimes result in a loss of 1 pixel.
            b=round((j+1)*h0)
            c=round(i*w0)
            d=round((i+1)*w0)
            img0=img[a:b,c:d]
            cv2.imencode('.png', img0)[1].tofile(img_names)
            i+=1
        j+=1
    


for i in file_name2('path_to_the_images'):#
    if "the Special characters used for identifying target images"in i:#############
        print(i)
        divide_spot(i)
```

```python
%%time
########################################
######################################################
##################Fill in the empty slots that do not contain mutant data, arranging the groups of data in the order from A1 to P24.##########
#
import numpy as np
from scipy import stats
f3=open('g_fhsvuniqname.csv','r')
f2=open('gsize.csv','w')

def changeseq_384(num):
    a=(num-1)//24
    tail=(num-1)%24+1
    head=chr(65+a)
    body=head+str(tail)
    return body
def dictseq_0(dictseq):
    i=1
    while i<385:
        dictseq[changeseq_384(i)]=changeseq_384(i)+",NA"*1
        i+=1
    return dictseq
dictseq={}
b=f3.readline()
f2.write("plate,"+",".join(b.split(",")[:2])+"\n")
while 1:
    a=f3.readline().strip()
    if a=="":break
    b=a.split(",")
    if len(b[0].split("-"))==3:
    
        name=b[0].split("-")[0]+"-"+b[0].split("-")[1]
    elif len(b[0].split("-"))==2:
        name=b[0].split("-")[0]
    else:print(a)
        
    site=b[0].split("-")[-1]
    if name in dictseq: 
        dictseq[name][site]=a
    else:
        dictseq1={}
        dictseq[name]=dictseq_0(dictseq1)
        dictseq[name][site]=a
    
for key in sorted(dictseq):
    for keys in dictseq[key]:
        if "NA" in dictseq[key][keys]:
            f2.write(key+","+key+"-"+dictseq[key][keys]+"\n")
        else:
            f2.write(key+","+",".join(dictseq[key][keys].split(",")[:2])+"\n")
f2.close()
f3.close()
print("The end!")
```
